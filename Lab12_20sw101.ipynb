```python
import requests

# Fetch a web page
r = requests.get("https://medium.com/@nayabmemon2002/adds-analysis-33a8e9d12ea7")
print(r.text)
```

```python
import re

# Remove HTML tags using RegEx
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))
```

```python
from bs4 import BeautifulSoup

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python

# Find all articles
summaries = soup.find_all("p")
summaries[0]
```

```python
#Sample text
text = "HEY YOU ARE LOOKING GOOD. It will change your view of the matrix. Are the human people the ones who started the war ? Is AI a bad thing ?"
print(text)
```

```python
 #Convert to lowercase
text = text.lower()
print(text)
```

```python
#Punctuation Removal
import re

# Remove punctuation characters
text = re.sub(r"[^a-zA-Z0-9]", " ", text)
print(text)
```

```python
#Tokenization
# Split text into tokens (words)
words = text.split()
print(words)
```

```python
#NLTK: Natural Language ToolKit
import os
import nltk
nltk.download('punkt')
nltk.data.path.append(os.path.join(os.getcwd(), "nltk_data"))
```

```python
#Another sample text
text = "Dr. Smith graduated from the University of Washington. He later started an analytics firm called Lux, which catered to enterprise customers."
print(text)
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
```

```python

# Reduce words to their stems
stemmed = [PorterStemmer().stem(w) for w in words]
print(stemmed)
```

```python

#Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer

```

```python
from nltk.tokenize import word_tokenize

# Split text into words using NLTK
words = word_tokenize(text)
print(words)
```
